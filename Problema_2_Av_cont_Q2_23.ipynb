```python
import cmath 
import numpy as np
from IPython.display import clear_output
import matplotlib.pyplot as plt
```

```python
nt=1.0;
ni=1.5;
num_punts=200;
thetai=np.linspace(0.01, np.pi/2, num=num_punts);
thetat=np.zeros((len(thetai),1),dtype=complex)
for i in range(len(thetai)):
    thetat[i,0]=(cmath.asin(ni*cmath.sin(thetai[i])/nt));
    #thetat[i,i]=thetat_temp
```

```python
#Fresnel Coefficients
rs=np.zeros((len(thetai),1),dtype=complex)
rp=np.zeros((len(thetai),1),dtype=complex)
for i in range(len(thetai)):
    rs[i,0]=cmath.sin(thetat[i,0]-thetai[i])/cmath.sin(thetai[i]+thetat[i,0]);
    rp[i,0]=cmath.tan(thetat[i,0]-thetai[i])/cmath.tan(thetai[i]+thetat[i,0]);
```

```python
plt.plot(thetai*180/np.pi,np.abs(rp),'b')
plt.plot(thetai*180/np.pi,np.abs(rs),'k')
```

```python
# modulus rp and rs
rp_mod=abs(rp);
rs_mod=abs(rs);

# phases rp and rs
phase_p=np.zeros((len(thetai),1),dtype=int)
phase_s=np.zeros((len(thetai),1),dtype=int)
for i in range(len(thetai)):
    #phase_p[i,0]=cmath.phase(rp[i,0])*180/np.pi;
    #phase_s[i,0]=cmath.phase(rs[i,0])*180/np.pi;
    #phase_p[i,0]=np.arctan2(np.imag(rp[i,0]), np.real(rp[i,0]))*180/np.pi;
    #phase_s[i,0]=np.arctan2(np.imag(rs[i,0]), np.real(rs[i,0]))*180/np.pi;
    phase_p[i,0]=np.angle(rp[i,0])*180/np.pi;
    phase_s[i,0]=np.angle(rs[i,0])*180/np.pi;

phase_rel=phase_s-phase_p;
```

```python
plt.plot(thetai*180/np.pi,phase_p,'b')
plt.plot(thetai*180/np.pi,phase_s,'k')
plt.plot(thetai*180/np.pi,phase_rel,'r')
```

```python
A=1;
Ax=A;
Ay=A;

phase_rel_a=-45*np.pi/180;
w=1;
num_punts=100;
t=np.linspace(0, 2*np.pi/w, num=num_punts);

Ex=Ax*np.cos(w*t);
Ey=Ay*np.cos(w*t+phase_rel_a);
clear_output(wait=True)

for j in range(len(t)):
    clear_output(wait=True)
    plt.figure(num=1,figsize=(5, 5))
    plt.plot(-1*Ex[0:j],Ey[0:j], 'ro')
    plt.xlim(-1.1,1.1)
    plt.ylim(-1.1,1.1)
    #time.sleep(0.01)
    plt.show()
    
    

```
